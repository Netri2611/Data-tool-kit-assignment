### Q1: How do you create a 2D NumPy array and calculate the sum of each row?

```python
import numpy as np

array = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(array, axis=1)
print(row_sums)
```

### Q2: Write a Pandas script to find the mean of a specific column in a DataFrame.

```python
import pandas as pd

df = pd.DataFrame({'Marks': [85, 90, 78, 92]})
mean_marks = df['Marks'].mean()
print(mean_marks)
```

### Q3: Create a scatter plot using Matplotlib.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

### Q4: How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.DataFrame(np.random.rand(5, 3), columns=['A', 'B', 'C'])
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
```

Theory Answers


**Q1: What is NumPy, and why is it widely used in Python?**
**A:**
NumPy (Numerical Python) is a fundamental package for scientific computing in Python. It provides support for large, multi-dimensional arrays and matrices, along with a collection of high-level mathematical functions to operate on these arrays efficiently. It is widely used for numerical data processing due to its performance and ease of integration with other libraries.

---

**Q2: How does broadcasting work in NumPy?**
**A:**
Broadcasting in NumPy allows arrays of different shapes to be used in arithmetic operations. NumPy automatically expands the smaller array to match the shape of the larger array without copying data, making element-wise operations possible even when shapes differ, under specific rules.

---

**Q3: What is a Pandas DataFrame?**
**A:**
A Pandas DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table and provides powerful tools for data manipulation, analysis, and cleaning.

---

**Q4: Explain the use of the `groupby()` method in Pandas.**
**A:**
The `groupby()` method in Pandas is used to split the data into groups based on some criteria, apply a function to each group, and combine the results. It is commonly used for aggregation operations like sum, mean, or count.

---

**Q5: Why is Seaborn preferred for statistical visualizations?**
**A:**
Seaborn is preferred for statistical visualizations because it provides high-level interfaces for drawing attractive and informative statistical graphics. It integrates well with Pandas and simplifies complex visualizations like violin plots, box plots, and heatmaps.

---

**Q6: What are the differences between NumPy arrays and Python lists?**
**A:**

* NumPy arrays are more memory-efficient and faster for numerical operations.
* They support vectorized operations, unlike Python lists.
* NumPy arrays require elements to be of the same data type, whereas Python lists can hold mixed types.

---

**Q7: What is a heatmap, and when should it be used?**
**A:**
A heatmap is a graphical representation of data where values are represented by color. It is especially useful for visualizing the correlation matrix or any 2D matrix data to quickly identify patterns or relationships.

---

**Q8: What does the term “vectorized operation” mean in NumPy?**
**A:**
Vectorized operation means applying an operation to an entire array or a set of arrays without the need for explicit loops. This leads to faster computation and more concise code.

---

**Q9: How does Matplotlib differ from Plotly?**
**A:**

* **Matplotlib** is a static plotting library ideal for basic and publication-quality visualizations.
* **Plotly** is an interactive plotting library, better for web-based dashboards and complex visualizations.
  Plotly supports interactivity like zoom, hover, and real-time updates.

---

**Q10: What is the significance of hierarchical indexing in Pandas?**
**A:**
Hierarchical indexing allows a DataFrame or Series to have multiple index levels. This makes it easier to work with higher-dimensional data in a lower-dimensional form and enables more complex groupings and data manipulation.

---

**Q11: What is the role of Seaborn’s `pairplot()` function?**
**A:**
`pairplot()` creates a grid of scatter plots for each pair of features in a dataset. It’s useful for exploring relationships between variables and understanding data distributions, especially in exploratory data analysis (EDA).

---

**Q12: What is the purpose of the `describe()` function in Pandas?**
**A:**
The `describe()` function generates descriptive statistics of a DataFrame, such as mean, median, standard deviation, min, max, and quartiles. It provides a quick overview of the distribution and spread of numerical data.

**Q15: How does NumPy handle multidimensional arrays?**
**A:**
NumPy supports n-dimensional arrays (ndarrays) and provides functions to create, reshape, slice, and perform operations on them efficiently. It allows mathematical operations across multiple axes and dimensions using vectorization.

**Q16: What is the role of Bokeh in data visualization?**

**Q17: Explain the difference between `apply()` and `map()` in Pandas.**
**A:**

* `map()` is used for element-wise transformations on Series.
* `apply()` is more flexible and can be used on Series and DataFrames; it can apply a function along an axis (rows or columns).
  Use `map()` for simple value substitutions and `apply()` for complex row or column-wise operations.



**Q18: What are some advanced features of NumPy?**
**A:**
Some advanced features of NumPy include:

* **Broadcasting:** For efficient computation on arrays of different shapes.
* **Masked arrays:** For handling invalid or missing data.
* **Structured arrays:** For heterogeneous data (similar to SQL table rows).
* **Memory-mapped arrays:** For reading large binary data files without loading the entire file into memory.
* **Advanced indexing and slicing:** For selecting complex subsets of data.
* **Universal functions (ufuncs):** For fast, element-wise operations with broadcasting support.

---

**Q19: How does Pandas simplify time series analysis?**
**A:**
Pandas simplifies time series analysis by offering:

* Built-in datetime objects and indexing (`DatetimeIndex`)
* Time-based indexing and slicing
* Resampling and frequency conversion
* Rolling window calculations (e.g., moving average)
* Time zone handling
* Convenient functions like `.shift()`, `.diff()`, `.asfreq()` for temporal operations

---

**Q20: What is the role of a pivot table in Pandas?**
**A:**
A pivot table in Pandas is used to reshape and summarize data. It helps in:

* Aggregating data across multiple dimensions
* Performing grouping and applying statistical functions (e.g., mean, sum)
* Converting long-format data into wide format for easier analysis

---

**Q21: Why is NumPy’s array slicing faster than Python’s list slicing?**
**A:**
NumPy’s array slicing is faster because:

* It returns views instead of copies, avoiding unnecessary memory allocation
* Arrays are stored in contiguous memory blocks, making access faster
* NumPy is implemented in C under the hood, which optimizes performance

---

**Q22: What are some common use cases for Seaborn?**
**A:**
Common use cases for Seaborn include:

* Visualizing distributions (histograms, KDE plots)
* Comparing categories (box plots, bar plots)
* Analyzing relationships (scatter plots, pair plots, regression plots)
* Heatmaps for correlation matrices or pivot tables
* Time series and trend visualization with less code and better aesthetics
